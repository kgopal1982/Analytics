```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('c:\\Datasets\\Association_Spend_Data.csv')
df.head()
```

```python
df.info()
```

```python
del df['OrderId']
del df['Sex']
del df['Age']
```

```python
df.head()
```

```python
#reference is http://pbpython.com/market-basket-analysis.html
```

```python
#generate frequency item sets
```

```python
frequent_itemsets = apriori(df, min_support=0.07, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(10)
```

```python
rules[ (rules['lift'] >= 4) &
       (rules['confidence'] >= 0.8) ]
```
