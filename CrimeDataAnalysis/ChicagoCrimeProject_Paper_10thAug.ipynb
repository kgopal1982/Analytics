```python
crimes_2017_3 = crimes_2017_2.copy(deep=True)
```

```python
crimes_2017_3['Ward']= crimes_2017_3['Ward'].astype(int)
```

```python
crimes_2017_3.info()
```

```python
crimes_2017_3.head(2)
```

```python
crimes_2017_3['Ward']= crimes_2017_3['Ward'].astype(object)
crimes_2017_3['Beat']= crimes_2017_3['Beat'].astype(object)
crimes_2017_3['Quarter']= crimes_2017_3['Quarter'].astype(object)
```

```python
crimes_2017_3.info()
```

```python
cross1 = pd.crosstab(crimes_2017_3['Primary Type'], crimes_2017_3['Location Description'])
```

```python
from scipy.stats import chi2_contingency
```

```python
chi2_contingency(cross1)
```

```python
#As p value is less then 0.05, hence there is relationship between Primary Type and Location Description
```

```python
cross2 = pd.crosstab(crimes_2017_3['Primary Type'], crimes_2017_3['Ward'])
```

```python
chi2_contingency(cross2)
```

```python
#As p value is less then 0.05, hence there is relationship between Primary Type and Ward
```

```python
cross3 = pd.crosstab(crimes_2017_3['Primary Type'], crimes_2017_3['Beat'])
```

```python
chi2_contingency(cross3)
```

```python
#As p value is less then 0.05, hence there is relationship between Primary Type and Beat
```

```python
#Check Stationarity by plotting Rolling Statistics and Dickey-Fuller Test
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.xticks(rotation='vertical')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
```
