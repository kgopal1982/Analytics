```python
crimes_2017_theft_df.head()
```

```python
X_theft=crimes_2017_theft_df.iloc[:, [0,1]].values
```

```python
type(X_theft)
```

```python
#using the elbow method to find the optimal no of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10,random_state=0)
    kmeans.fit(X_theft)
    wcss.append(kmeans.inertia_) #it calculates within clusters sum of squares
plt.plot(range(1,11),wcss)
plt.title('The ELblow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#cluster will be 3
```

```python
#applying k-mean to the dataset
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10,random_state=0)
y_kmeans = kmeans.fit_predict(X_theft)
```

```python
y_kmeans
```

```python
#Visualize the clusters
plt.scatter(X_theft[y_kmeans == 0,0], X_theft[y_kmeans == 0,1],s =100, c = 'red', label = 'Cluster 1')
plt.scatter(X_theft[y_kmeans == 1,0], X_theft[y_kmeans == 1,1],s =100, c = 'blue', label = 'Cluster 2')
plt.scatter(X_theft[y_kmeans == 2,0], X_theft[y_kmeans == 2,1],s =100, c = 'green', label = 'Cluster 3')
#plt.scatter(X_theft[y_kmeans == 3,0], X_theft[y_kmeans == 3,1],s =100, c = 'cyan', label = 'Cluster 4')
#plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s=300, c='yellow',label='Centroid')
plt.title("Clustering- No of Thefts across Wards in 2017")
plt.xlabel("Ward")
plt.ylabel("Crime Counts")
plt.legend()
plt.show()
```

# USE FB Prophet model for Forecasting

```python
from fbprophet import Prophet
```

```python
ts = monthly_theft
```

```python
ts
```

```python
fd = ts.copy(deep=True)
```

```python
fd.head()
```

```python
fd_df = pd.Series.to_frame(fd)
```

```python
fd_df.head()
```

```python
fd_df['Months'] = fd_df.index
```

```python
fd_df.head()
```

```python
fd_df.reset_index(inplace=True)
```

```python
fd_df.head()
```

```python
fd_df =fd_df[['Date', 'Months']].copy()
```

```python
fd_df.head()
```

```python
fd_df = fd_df.rename(columns={"Date":"y","Months":"ds"})
```

```python
fd_df=fd_df.sort_index(axis=1)
```

```python
fd_df.head()
```

```python
#Build timeseries model
```

```python
my_forecast = Prophet(interval_width=0.95)
```

```python
my_forecast.fit(fd_df)
```

```python
forecast_periods = my_forecast.make_future_dataframe(periods=36, include_history=True, freq='M')
forecast_periods
```

```python
forecast = my_forecast.predict(forecast_periods)
forecast.head()
```

```python
forecast1 = forecast.loc[:,["ds", "yhat"]]
```

```python
forecast1.head()
```

```python
my_forecast.plot(forecast,uncertainty=True);
```

```python
my_forecast.plot_components(forecast)
plt.show();
```

```python
type(forecast1)
```

```python
final_df = pd.merge(forecast1, fd_df, on = 'ds', how='inner')
```

```python
final_df.head()
```

```python
#compute accuracy of the forecast
```

```python
import numpy as np
from sklearn import metrics
```

```python
np.round(metrics.mean_absolute_error(final_df.yhat,final_df.y),2)
```
