```python
#load the dataset
iris=sns.load_dataset("iris")
```

### Analyze distribution of the series using Histogram

```python
#Histogram and Kernel Density Estimator
sns.distplot(iris.sepal_length)
plt.show();
```

```python
sns.distplot(iris.sepal_width)
plt.show();
```

```python
sns.distplot(iris.petal_length)
plt.show();
```

```python
sns.distplot(iris.petal_width)
plt.show();
```

### Analyze distribution of the series using PairPlot

```python
sns.pairplot(iris, hue = "species", size =5, x_vars = ["sepal_width", "sepal_length"], y_vars = ["petal_width", "petal_length"])
plt.show();
```

### Analyze distribution of the series using Box Plot

```python
#Box plot
#plt.figure(figsize=(10,15))
sns.boxplot(x='species', y='sepal_length', data = iris, linewidth = 2)
plt.show();
```

```python
sns.boxplot(x='species', y='sepal_width', data = iris, linewidth = 2)
plt.show();
```

```python
sns.boxplot(x='species', y='petal_length', data = iris, linewidth = 2)
plt.show();
```
